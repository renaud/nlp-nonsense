```python
%pylab inline
import pandas as pd
import sys, os, re, json
from collections import Counter
```

```python
# Load annotations
files = ["data/annotations-110214/commoncrawl-sample.filtered.part1.txt.annotated",
         "data/annotations-110214/commoncrawl-sample.filtered.part2.txt.annotated"]
data = {}
for fname in files:
    with open(fname) as f:
        data.update(json.load(f))
        
df = pd.DataFrame.from_records(data.items(), columns=['text','label'])

def remap_to_tag(l):
    if l.startswith('-'): return l # idempotent
    if l.startswith('Sentence'): return "-SENTENCE-"
    if l.startswith('Nonsense'): return "-NONSENSE-"
    if l.startswith('Porn'): return "-PORN-"
    if l.startswith('No'): return "-NONE-"
    return None

df['label'] = df.label.map(remap_to_tag)
df.head(10)
```

```python
Counter(df.label)
```

```python
# Compute statistics
def count_nonalpha(a):
    return len([c for c in a if (not c.isalpha())])

df['nchars'] = df.text.map(len)
df['nwords'] = df.text.map(lambda l: len(l.split()))
df['nnonalpha'] = df.text.map(count_nonalpha)
```
