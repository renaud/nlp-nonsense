```python
import bokeh.plotting as bp
from bokeh.models import HoverTool
from collections import OrderedDict
bp.output_notebook()

colors_rgb = ["#006400", "#00008b", "#8b0000", "#696969"]
colorlist = [colors_rgb[i] for i in y]
source = bp.ColumnDataSource(
    data=dict(x=Xt[:,0], y=Xt[:,1], color=colorlist, 
              gloss=list(dataset.df_master.__TEXT__),
              label=list(dataset.df_master.__LABEL__))
)

bp.output_file('viz.html')

TOOLS = "resize,hover,pan,wheel_zoom,reset,save"

p = bp.figure(title="t-SNE visualization", tools=TOOLS)
p.circle("x", "y", 1, 1, source=source, 
         color="color", alpha=0.3, size=7)
bp.hold()

hover = p.select(dict(type=HoverTool))
hover.snap_to_data = True
hover.tooltips = OrderedDict([
    ('text', '@gloss'),
    ('label', '@label'),
])

bp.show(p)
bp.save(obj=p)
```
